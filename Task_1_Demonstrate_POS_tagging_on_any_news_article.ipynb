```
import pandas as pd

train = pd.read_csv('news-topic-classification/BBC News Train.csv')
train.head()
```

```
import nltk

nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('tagsets')
nltk.download('brown')
nltk.download('universal_tagset')
```

```
import nltk
from nltk.tag import pos_tag, map_tag

def adjective(row):
  text = nltk.word_tokenize(row)
  posTagged = pos_tag(text)
  simplifiedTags = [(word, map_tag('en-ptb', 'universal', tag)) for word, tag in posTagged]
  
  adj = []
  for x in simplifiedTags:
    if x[1] == 'ADJ':
      adj.append(x[0])

  return adj

def article(row):
  text = nltk.word_tokenize(row)
  posTagged = pos_tag(text)
  simplifiedTags = [(word, map_tag('en-ptb', 'universal', tag)) for word, tag in posTagged]
  
  det = []
  for x in simplifiedTags:
    if x[1] == 'DET':
      det.append(x[0])

  return det

def conjuction(row):
  text = nltk.word_tokenize(row)
  posTagged = pos_tag(text)
  simplifiedTags = [(word, map_tag('en-ptb', 'universal', tag)) for word, tag in posTagged]
  
  conj = []
  for x in simplifiedTags:
    if x[1] == 'CONJ':
      conj.append(x[0])

  return conj

def noun(row):
  text = nltk.word_tokenize(row)
  posTagged = pos_tag(text)
  simplifiedTags = [(word, map_tag('en-ptb', 'universal', tag)) for word, tag in posTagged]
  
  nn = []
  for x in simplifiedTags:
    if x[1] == 'NOUN':
      nn.append(x[0])

  return nn

def preposition(row):
  text = nltk.word_tokenize(row)
  posTagged = pos_tag(text)
  simplifiedTags = [(word, map_tag('en-ptb', 'universal', tag)) for word, tag in posTagged]
  
  pre = []
  for x in simplifiedTags:
    if x[1] == 'ADP':
      pre.append(x[0])

  return pre

def pronoun(row):
  text = nltk.word_tokenize(row)
  posTagged = pos_tag(text)
  simplifiedTags = [(word, map_tag('en-ptb', 'universal', tag)) for word, tag in posTagged]
  
  pron = []
  for x in simplifiedTags:
    if x[1] == 'PRON':
      pron.append(x[0])

  return pron

def verb(row):
  text = nltk.word_tokenize(row)
  posTagged = pos_tag(text)
  simplifiedTags = [(word, map_tag('en-ptb', 'universal', tag)) for word, tag in posTagged]
  
  vrb = []
  for x in simplifiedTags:
    if x[1] == 'VERB':
      vrb.append(x[0])

  return vrb

def adverb(row):
  text = nltk.word_tokenize(row)
  posTagged = pos_tag(text)
  simplifiedTags = [(word, map_tag('en-ptb', 'universal', tag)) for word, tag in posTagged]
  
  adv = []
  for x in simplifiedTags:
    if x[1] == 'ADV':
      adv.append(x[0])

  return adv

def count(df):
  adjective = len(df['adjective'])
  article = len(df['article'])
  conjuction = len(df['conjuction'])
  noun = len(df['noun'])
  preposition = len(df['preposition'])
  pronoun = len(df['pronoun'])
  verb = len(df['verb'])
  adverb = len(df['adverb'])

  return 'adjective : {0}, article: {1}, conjuction: {2}, noun: {3}, preposition: {4}, pronoun: {5}, adverb : {6}, verb : {7}'.format(adjective, article, conjuction, noun, preposition, pronoun, adverb, verb)
```

```
train['adjective'] = train['Text'].map(lambda x: adjective(x))
train.head()
```

```
train['article'] = train['Text'].map(lambda x: article(x))
train.head()
```

```
train['conjuction'] = train['Text'].map(lambda x: conjuction(x))
train.head()
```

```
train['noun'] = train['Text'].map(lambda x: noun(x))
train.head()
```

```
train['preposition'] = train['Text'].map(lambda x: preposition(x))
train.head()
```

```
train['pronoun'] = train['Text'].map(lambda x: pronoun(x))
train.head()
```

```
train['verb'] = train['Text'].map(lambda x: verb(x))
train.head()
```

```
train['adverb'] = train['Text'].map(lambda x: adverb(x))
train.head()
```

```
train['count'] = train.apply(count, axis=1)
```
